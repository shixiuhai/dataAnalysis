```python
# -*- coding: utf-8 -*-
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn import metrics
 
# 假如我要构造一个聚类数为2的聚类器
estimator = KMeans(n_clusters=2)#构造聚类器
kmeans_data_tf=np.array([[1,1],[2,2],[3,1],[6,4],[7,5],[8,4]])

estimator.fit(kmeans_data_tf) #聚类
 
r2 = pd.DataFrame(estimator.cluster_centers_)  #找出聚类中心

r3=metrics.silhouette_score(kmeans_data_tf, estimator.labels_, metric='euclidean') # 轮廓系数
print("聚类中心是%s\n轮廓系数是%s"%(r2,r3))
```
